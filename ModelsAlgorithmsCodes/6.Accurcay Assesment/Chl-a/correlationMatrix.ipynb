```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#import pymc3 as pm

from matplotlib import animation, rc
%matplotlib inline
```

```python
df = pd.read_csv('Data.csv')
```

```python
data=df[['column1',	'column2', 'column3', 'column_n']]
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(), annot=True, linewidth=1.0, cmap='Reds')
plt.title('CORRELATION BETWEEN FACTORS AFFECTING MALARIA AND OCCURANCES')
plt.savefig('Correlation.png',dpi=1000, bbox_inches='tight')
plt.show()
```
